# Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`. 

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# PART 4:

import numpy as np

# Assuming 'img_average_channel' is already defined

# Get the number of rows in 'img_average_channel'
num_rows = img_average_channel.shape[0]

# Create an array containing numbers 0 to N-1
row_indices = np.arange(num_rows)

# Shuffle the row indices using numpy.random.shuffle
np.random.shuffle(row_indices)

# Print the shuffled row indices
print("Shuffled Row Indices:")
print(row_indices)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
import numpy as np

# Task 1: Create matrix T using the shuffled row indices
T = identity_matrix[row_indices]

# Task 2: Describe what matrix T represents
# Matrix T is a reordering of the identity matrix where the rows have been shuffled according to the 'row_indices'.

# Task 3: Sanity check - Print the shape of T
print("Shape of T:", T.shape)

# We will Reshape T if shapes of T and identity_matrix don't match
if T.shape != identity_matrix.shape:
    T = T.reshape(identity_matrix.shape)
    print(f"Reshaped T: {T.shape}")

```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Matrix multiply 'T' and 'img_average_channel' to get 'transformed_image'
transformed_image = T @ img_average_channel

# Check the shape of the result
print("Shape of transformed_image:", transformed_image.shape)

# Plot 'transformed_image'
plt.imshow(transformed_image)
plt.title("Transformed Image")
plt.show()

# Task 3: Describe what the transformation is doing to the original image
# 'T @ img_average_channel' rearranges the rows of 'img_average_channel' using a pattern defined by 'T'. 
# Each row in 'transformed_image' corresponds to a row from 'img_average_channel', but their order is shuffled based on 'T'.
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
import numpy as np

# Check if T is a square matrix (same number of rows and columns)
if T.shape[0] == T.shape[1]:
    # Calculate the determinant of T
    det_T = np.linalg.det(T)
    
    # Check if the determinant is non-zero
    if det_T != 0:
        # If T is square and invertible, compute its inverse
        Tinv = np.linalg.inv(T)
        
        # Use the inverse to retrieve the original image
        original_image = Tinv @ transformed_image
        
        # Plot the retrieved original image
        plt.imshow(original_image)
        plt.title('Retrieved Original Image')
        plt.show()
    else:
        print("Matrix T is not invertible (determinant is zero).")
else:
    print("Matrix T is not square (not invertible).")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
import numpy as np
import matplotlib.pyplot as plt


# Get the number of rows and columns in the image
num_rows, num_columns, _ = image.shape

# Create random permutations of row and column indices
rows_permutation = np.random.permutation(num_rows)
cols_permutation = np.random.permutation(num_columns)

# Shuffle both rows and columns of the original image
image_scrambled_both = image[rows_permutation][:, cols_permutation]

# Plot the scrambled image with both row and column scrambling
plt.imshow(image_scrambled_both)
plt.title('Scrambled Image (Rows and Columns)')
plt.show()
```

```python
# Now, invert the operations to retrieve the original image
# First, unshuffle the columns (undo column scrambling)
image_unscrambled_cols = image_scrambled_both[:, np.argsort(cols_permutation)]

# Next, unshuffle the rows (undo row scrambling)
image_unscrambled = image_unscrambled_cols[np.argsort(rows_permutation)]

# Plot the retrieved original image
plt.imshow(image_unscrambled)
plt.title('Retrieved Original Image')
plt.show()
```
